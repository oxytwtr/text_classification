## Описание задачи

*Исходные данные от заказчика*

- Данные лежат в файле `data.csv`. 
- Столбец text в нём содержит текст комментария, а class — целевой признак.

*Задача*

Необходимо построить модель для классификации отзывов клиентов.
Метрика для оценки - f1

### План работы

1. Загрузить данные
2. Очистить пропуски, провести токенизацию и лемматизацию текстов.
3. Выбрать и создать признаки, предполагающиеся для исследования
4. Выбрать модели, предполагающиеся для исследования, обосновать
5. Выделить тестовую выборку размером 10% от исходных данных.
6. Оценить важность признаков средствами sklearn
7. Обучить разные модели с различными гиперпараметрами. 
8. Проверить лучшую модель на тестовой выборке и сделать выводы.
9. Сохранить модель
10. Создать скрипт `model.py` с функцией `get_result(text: pd.Series) -> pd.Series:`

````python
def get_direct_path_yadisk(folder_url, filename):
    '''
    makes a direct link (for download) to the file located in the Yandex disk public link
    '''
    url = ('https://cloud-api.yandex.net/v1/disk/public/resources/download' +
           '?public_key=' + urllib.parse.quote(folder_url) + '&path=/' + 
           urllib.parse.quote(filename))
    r = requests.get(url) 
    return json.loads(r.text)['href']    
````

````python
def label_enc(list_with_classes_):
    '''
    series_with_classes_: List-like, contains the values of a categorical feature
    The function encodes categorical features by direct encoding,
    returns:
        labels_: list of value labels
        dict_ as format: {label:value, ...}
        le_: ftted encoder
    '''
    le_ = LabelEncoder()
    labels_ = le_.fit_transform(list_with_classes_)
    dict_ = {i: list(le_.classes_)[i] for i in range(0, len(list(le_.classes_)))}
    return labels_, dict_, le_
````

````python
def merge_texts_in_list(corpus_list_, req_count) -> list:
    '''
    corpus_list_: list-like, that contains the texts to be converted
    req_count: int, the minimum number of tokens in the text after conversion
    
    The function accepts a list with texts and joins the texts in it 
    so that the resulting text contains at least req_count tokens
    returns list with transformed texts
    '''
 
    new_corpus_list_ = []
    temp_text_ = ''
    
    for text_ in corpus_list_:
        count_ = len(re.findall(r'(?u)\b\w\w+\b', temp_text_))  

        if count_ >= req_count:
            temp_text_ = temp_text_.strip()
            new_corpus_list_.append(temp_text_)
            temp_text_ = text_
        else:
            text_ = text_.strip()
            temp_text_ += text_ + ' '
            
    return new_corpus_list_
````

````python
def split_texts_in_list(corpus_list_, req_count) -> list:
    '''
    corpus_list_: list-like, that contains the texts to be converted
    req_count: int, the minimum number of tokens in the text after conversion
    
    The function accepts a list with texts and splits the texts in it 
    so that the resulting text contains at least req_count tokens
    returns list with transformed texts
    '''
    
    new_corpus_list_ = []
        
    for text_ in corpus_list_:
        text_ = text_.split()
                
        for j in range(0, len(text_), req_count):
            new_corpus_list_.append(
                ' '.join(text_[j:min(j + req_count, len(text_))])
            )
                        
    return new_corpus_list_
````

````python
def merge_texts_in_df(corpus_df_, text_column_name, count_column_name, req_count) -> pd.DataFrame:
    '''
    corpus_df_: pd.DataFrame, that contains the texts to be converted
    text_column_name: str, name of the column that contains the texts
    count_column_name: str, name of the column that contains the number of tokens in the text
    req_count: int, the minimum number of tokens in the text after conversion
    
    The function accepts a pd.DataFrame with texts and joins the texts in it 
    so that the resulting text contains at least N tokens
    returns transformed pd.DataFrame
    '''
    
    new_corpus_df_ = pd.DataFrame(columns=corpus_df_.columns)
    temp_series_ = pd.Series(index=corpus_df_.columns)
    temp_series_[text_column_name] = ''
    temp_series_[count_column_name] = 0
    other_columns_ = list(corpus_df_.columns)
    other_columns_.remove(text_column_name)
    other_columns_.remove(count_column_name)
    
    for i, row in corpus_df_.iterrows():
                
        if temp_series_[count_column_name] >= req_count:
            new_corpus_df_ = new_corpus_df_.append(temp_series_, ignore_index=True)
            temp_series_ = row
        else:
            temp_series_[other_columns_] = row[other_columns_]
            temp_series_[text_column_name] = str(temp_series_[text_column_name]) + str(row[text_column_name])
            temp_series_[count_column_name] = temp_series_[count_column_name] + row[count_column_name]
            
    new_corpus_df_ = new_corpus_df_.astype('int', errors='ignore')
    return new_corpus_df_
````

````python
def split_texts_in_df(corpus_df_, text_column_name, count_column_name, req_count) -> pd.DataFrame:
    '''
    corpus_df_: pd.DataFrame, that contains the texts to be converted
    text_column_name: str, name of the column that contains the texts
    count_column_name: str, name of the column that contains the number of tokens in the text
    req_count: int, the minimum number of tokens in the text after conversion
    
    The function accepts a pd.DataFrame with texts and splits the texts in it 
    so that the resulting text contains at least N tokens
    returns transformed pd.DataFrame
    '''
    
    new_corpus_df_ = pd.DataFrame(columns=corpus_df_.columns)

    other_columns_ = list(corpus_df_.columns)
    other_columns_.remove(text_column_name)
        
    for i, row in corpus_df_.iterrows():
        text_ = row[text_column_name].split()
        
        for j in range(0, len(text_), req_count):
                       
            row[count_column_name] = min(req_count, len(text_) - j)
            new_corpus_df_ = new_corpus_df_.append(
                pd.Series(' '.join(text_[j:min(j + req_count, len(text_))]),
                          index=[text_column_name]).append(
                    row[other_columns_]), ignore_index=True
                )
            
    new_corpus_df_ = new_corpus_df_.astype('int', errors='ignore')                    
    return new_corpus_df_
````

````python
def best_corr(dataset, range_=(0.5, 1)):
    '''function displays best correlation coefficients for pairs of features in corr matrix
    return pd.DataFrame with pairs of features and their correlation coefficients
    '''
    corr_table = pd.DataFrame({'first_feature': [],
                               'second_feature': [],
                               'corr_coef': []})
    exist_list = []
    for col_name, data in dataset.items():
        exist_list.append(col_name)
    
        for second in data.items():
            corr_value = round(second[1], 2)
            if corr_value > range_[0] and corr_value <= range_[1]:
                if second[0] not in exist_list:
                    corr_table = corr_table.append({'first_feature': col_name,
                                                    'second_feature': second[0],
                                                    'corr_coef': corr_value},
                                                   ignore_index=True)  
                
    return corr_table.sort_values(by='corr_coef', ascending=False, ignore_index=True)
````

````python
def single_feature_comp(corr_matrix_, target_feat_name, threshold_, direction_='high'):
    '''
    corr_matrix_: pd.DataFrame with feature correlation estimates
    target_feat_name: str, name of the column that contains target feature
    threshold: int, "high" or "low"
    return: if threshold = "high", returns a list with pairs whose correlation is above the threshold
            else returned list with pairs whose correlation is below the threshold
    '''
    if direction_ == 'high':
        return round(abs(
            corr_matrix_[target_feat_name][(abs(corr_matrix_[target_feat_name]) > threshold_) &
                                           (abs(corr_matrix_[target_feat_name]) != 1)]),
                     2).sort_values(ascending=False)
    return round(abs(
            corr_matrix_[target_feat_name][(abs(corr_matrix_[target_feat_name]) < threshold_) &
                                           (abs(corr_matrix_[target_feat_name]) != 1)]),
                     2).sort_values(ascending=False)
````

#### Токенизация и лемматизация текста

Токенизацию будем делать используя регулярные выражения, лемматизацию - библиотекой `pymystem3`.

Объявим функцию для токенизации и лемматизации текста и применим ее к корпусу с текстами

````python
def tokenize_lemmatize_pymystem(list_of_text, _stopwords):
    '''
    list_of_text: list-like, must contain text
    _stopwords: list-like, must contain text with stop words to exclude from the output set
    the function performs preprocessing of a list of texts using re, pymystem3
    
    Returns a list with preprocessed text
    Also shows the progress bar using tqdm
    '''
    
    m = Mystem()    
    return_series = []
    try:
        for text in tqdm(list_of_text):
            string = re.sub(r'[^а-яёА-ЯЁ\s]', '', text)
        
            string = m.lemmatize(string)
        
            string = ' '.join([token for token in string if token not in _stopwords])   
            string = re.sub(r'\s+', ' ', string, flags=re.I)
        
            return_series.append(string)
            
    except ImportError:
        for text in list_of_text:
            string = re.sub(r'[^а-яёА-ЯЁ\s]', '', text)
        
            string = m.lemmatize(string)
        
            string = ' '.join([token for token in string if token not in _stopwords])   
            string = re.sub(r'\s+', ' ', string, flags=re.I)
        
            return_series.append(string)
    
    return return_series
````

## Создание признаков

Исследованием предполагается создание и изучение следующего набора признаков, получаемых из исходных данных:

1. Группа 1 - CLASSIC, самостоятельно подготовленные классические признаки для классификации текста

   1. Признак: количество токенов в тексте. (проведем исследование корреляции с целевым)

   2. Признак наличия (в конкретном тексте из набора текстов) ключевого слова из определенного заранее списка. Предполагается категориальный бинарный признак, для каждого слова из списка (0/1 - слово есть в конкретном наборе или нет)

В тестовом наборе могут оказаться неизвестные кодировщику значения (комбинации ключевых слов), поэтому при объявлении кодировщика необходимо задать аргументы

````python
# adapted from https://www.sbert.net/examples/applications/computing-embeddings/README.html
# Mean Pooling - Take attention mask into account for correct averaging
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0]
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    sum_embeddings = torch.sum(token_embeddings * input_mask_expanded, 1)
    sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)
    return sum_embeddings / sum_mask
````

````python
# Объявление функции трансформации векторов в эмбеддинги загруженным трансформером

def get_embeddings(model_, dataloader_, device_=torch.device('cpu')):
    '''
    model_: trained transformer
    dataloader_: DataLoader with TensorDataset
    device_: default torch.device('cpu')
    
    Function transforms text vectors into embeddings
    
    Returns list of embeddings, with the number of elements equal to the number of batches шт loader
    '''
    model_.eval()
    embeddings_ = torch.Tensor().to(device)
       
    for step, batch in tqdm(enumerate(dataloader_), total = len(dataloader_)):
        batch = [r.to(device_) for r in batch]
        sent_id_, mask_ = batch
        
        with torch.no_grad():
            output = model_(sent_id_, token_type_ids=None, attention_mask=mask_)
            
            embeddings_ = torch.cat([embeddings_, mean_pooling(output, mask_)])
        
    embeddings_ = embeddings_.numpy()
             
    return embeddings_
````

#### Проверка методом SequentialFeatureSelection

Чтобы получить представление о важности признаков, примением следующие оценщики:

- LogisticRegressionCV
- RandomForestClassifier

На каждой итерации процесса SequentialFeatureSelection мы выбираем лучший признак для добавления к выбранным признакам на основе оценки перекрестной проверки. 

То есть мы начинаем с 0 признаков и выбираем лучший единственный признак с наивысшим баллом - метод forward.
Обратный метод - начиная со всех признаков, последовательно удалять неважные - метод backward.

Процедура повторяется до тех пор, пока мы не достигнем желаемого количества выбранных признаков.

````python
def sfs_research(model_, feature_names_, X_, y_, on_backward=True):
    '''
    '''
    sfs_forward = SequentialFeatureSelector(
        model_, n_features_to_select=4, direction="forward").fit(X_, y_)
    
    if on_backward:
        sfs_backward = SequentialFeatureSelector(
            model_, n_features_to_select=4, direction="backward").fit(X_, y_)

    return feature_names_[sfs_forward.get_support()], feature_names_[sfs_backward.get_support()]
````

````python
def optuna_process(function_, n_trials_= N_TRIALS, names_=names, direction_='maximize', verbose=False):
    '''
    '''
    sampler_ = optuna.samplers.TPESampler()
    pruner_ = optuna.pruners.HyperbandPruner()

    # direction="maximize", т.к. чем больше метрика тем лучше
    study = optuna.create_study(direction=direction_,
                                sampler=sampler_, pruner=pruner_,
                                study_name=names_[-2])
    try:
        study.optimize(function_, n_trials=n_trials_, callbacks=[check_threshold])
    except ThresholdExceeded:
        print(f'Выход по исключению, получено значение метрики:\
        {study.best_value} >= {METRIC_GOOD_LIMIT}')

    best_trial_ = study.best_trial
    if verbose:
        print(f'model: {names_[-1]} tuned by Optuna, results:')
        print(f'tuned param: {best_trial_.params}')
        print(f'F1 score value on train set: {best_trial_.value:.3}')
    return best_trial_
````

````python
def update_result_table(best_trial_=best_trial, names_=names, 
                        result_model_table_=result_model_table,
                        best_f1_=best_f1, model_=model):
    '''
    '''
    reg_exp=[]
    for idx in range(0, len(names)-2):
        nidx = idx-len(names)
        reg_exp.append(f'({names[nidx]}__)')
    reg_exp = '|'.join(reg_exp)

    row_ = {'model': names_[-1],
            'f1_train': best_trial_.value, 
            'train_time': (best_trial_.datetime_complete - best_trial_.datetime_start).total_seconds()}

    row_ = row_ | {re.sub(reg_exp, '', key, count=0): best_trial_.params[key] for key in best_trial_.params}

    result_model_table_ = result_model_table_.append(row_, ignore_index=True)  

    if best_f1_ < best_trial_.value:
        best_f1_ = best_trial_.value
        best_model_ = model_.set_params(**best_trial.params)
    else:
        best_f1_ = best_f1_
        best_model_ = best_model
    return best_f1_, best_model_, result_model_table_
````

#### Исследование NeuralNetwork - простая полносвязная нейронная сеть 

Построим и обучим простую нейронную сеть с одним скрытым полносвязным слоем. 

Активатор выходного нейрона - логистическая регрессия, адаптированная для мультиклассовой классификации (softmax).

В этой модели балансировку классов не применяем.

Для использования методов кросс-валидации от sklearn обернем созданную модель Sequential при помощи KerasClassifier от Scikeras для  использования в cross_vas_score.

````python
def create_model(hidden_layer_dim, meta):
    '''
    hidden_layer_dim: hidden layer size
    meta: is a special argument that will be handed a dict containing input metadata
    Function creates a simple NN (Sequential) model using TensorFlow.Keras
    Returns Sequential.model
    '''

    # Получение значений аргументов из словаря meta Scikeras 
    n_features_in_ = meta["n_features_in_"]
    X_shape_ = meta["X_shape_"]
    n_classes_ = meta["n_classes_"]
    
    # Инициализация
    model_seq = keras.models.Sequential()
    # Входной слой
    model_seq.add(keras.layers.Dense(n_features_in_, input_shape=X_shape_[1:]))
    model_seq.add(keras.layers.Activation("relu"))
    # Скрытый слой
    model_seq.add(keras.layers.Dense(hidden_layer_dim))
    model_seq.add(keras.layers.Activation("relu"))
    # Выходной слой
    model_seq.add(keras.layers.Dense(n_classes_))
    model_seq.add(keras.layers.Activation("softmax"))
    
    return model_seq
````

````python
# Пользовательские функции для расчета F1-меры

def recall(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall_keras = true_positives / (possible_positives + K.epsilon())
    return recall_keras

def precision(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision_keras = true_positives / (predicted_positives + K.epsilon())
    return precision_keras

def f1(y_true, y_pred):
    p = precision(y_true, y_pred)
    r = recall(y_true, y_pred)
    return 2 * ((p * r) / (p + r + K.epsilon()))
````

````python
def objective_lgb_bert(trial):    
    
    params = {
        'n_estimators': trial.suggest_int(
            'n_estimators', N_EST_GB_RANGE[0], N_EST_GB_RANGE[1], N_EST_GB_RANGE[2]),
        'max_depth': trial.suggest_int(
            'max_depth', MAX_DEPTH_GB_RANGE[0], MAX_DEPTH_GB_RANGE[1], MAX_DEPTH_GB_RANGE[2]),        
        'learning_rate': trial.suggest_categorical(
            'learning_rate', ETA_GB_LIST),
        'class_weight': trial.suggest_categorical(
            'class_weight', CLASS_WEIGHT_LIST),
        'min_child_samples': trial.suggest_int(
            'min_child_samples', MIN_CH_SMPL_RANGE[0], MIN_CH_SMPL_RANGE[1], MIN_CH_SMPL_RANGE[2]),
        'num_leaves': trial.suggest_int(
            'num_leaves', NUM_LEAVES_RANGE[0], NUM_LEAVES_RANGE[1], NUM_LEAVES_RANGE[2]),        
    }

    model.set_params(**params)
    
    scores_ = cross_val_score(model, X_train_bert, y_train_bert,  
        cv=CV_VALUE, scoring=SCORER)
    
    return (abs(scores_)).mean()
````

##### Токенизация и создание тензоров с признаками

````python
# Максимальная дина вектора признаков устанавливается 
# равной соответствующей длине вектораизатора группы 1 признаков

tokens_train = tokenizer.batch_encode_plus(
    text_train_bert.to_list(),
    max_length = max_feat,
    padding = 'max_length',
    truncation = True
)
tokens_val = tokenizer.batch_encode_plus(
    text_val_bert.to_list(),
    max_length = max_feat,
    padding = 'max_length',
    truncation = True
)

train_seq = torch.tensor(tokens_train['input_ids'])
train_mask = torch.tensor(tokens_train['attention_mask'])
train_y = torch.tensor(labels_train_bert.to_list())

val_seq = torch.tensor(tokens_val['input_ids'])
val_mask = torch.tensor(tokens_val['attention_mask'])
val_y = torch.tensor(labels_val_bert.to_list())


# Создание наборов данных для загрузчика
train_data = TensorDataset(train_seq, train_mask, train_y)
val_data =  TensorDataset(val_seq, val_mask, val_y)
````

````python
# Размер батча подобран в 128, т.к. данных не очень много
batch_size = 16

# Создание загрузчика данных в модель

train_sampler = RandomSampler(train_data)
train_dataloader = DataLoader(train_data, sampler = train_sampler, batch_size = batch_size)

val_sampler = SequentialSampler(val_data)
val_dataloader = DataLoader(val_data, sampler = val_sampler, batch_size = batch_size)
````

##### Настройка параметров модели

````python
param_optimizer = list(model.named_parameters())

no_decay = ['bias', 'gamma', 'beta']
optimizer_grouped_parameters = [
    {'params': [p for n, p in param_optimizer if not any(nd in n for nd in no_decay)],
     'weight_decay_rate': 0.01},
    {'params': [p for n, p in param_optimizer if any(nd in n for nd in no_decay)],
     'weight_decay_rate': 0.0}
]

optimizer = AdamW(optimizer_grouped_parameters, lr=2e-5)
````

````python
weights = torch.tensor(class_weights, dtype = torch.float)
weights = weights.to(device)
````

##### Обучение модели

````python
def train():
    '''
    Local function - doing model training
    '''
    model.train()
    total_loss = 0
    
    for step, batch in tqdm(enumerate(train_dataloader), total = len(train_dataloader)):
        
        batch = [r.to(device) for r in batch]
        sent_id, mask, labels = batch
        
        optimizer.zero_grad()
        
        output = model(sent_id, token_type_ids=None, attention_mask=mask, labels=labels)
        
        loss = output.loss
        total_loss += loss.item()
        
        loss.backward()
        
        optimizer.step()
                
    avg_loss = total_loss / len(train_dataloader)
       
    return avg_loss
````

````python
def evaluate():
    '''
    Local function - doing model validating
    '''
    model.eval()
    valid_preds = []
    valid_labels = []
    
    star_time = time.mktime(time.gmtime())

    for step, batch in tqdm(enumerate(val_dataloader), total = len(val_dataloader)):
        batch = [t.to(device) for t in batch]
        sent_id, mask, labels = batch
        
        with torch.no_grad():
            output = model(sent_id, mask)
       
        logits = output[0].numpy()
        label_ids = labels.numpy()
        
        batch_preds = np.argmax(logits, axis=1)
        
        batch_labels = label_ids
        valid_preds.extend(batch_preds)
        valid_labels.extend(batch_labels)
    
    end_time = time.mktime(time.gmtime())
        
    val_time = end_time - star_time
    
    return valid_preds, valid_labels, val_time
````

````python
epochs = 8
train_losses = []

print(f'Обучение модели {names[-1]}')

for epoch in range(epochs):
    print(f'\n - Epoch {epoch+1} / {epochs}')
    
    train_loss = train()
    train_losses.append(train_loss)
    
    print(f'\nTraining loss: {train_loss:.3f}')
````
